```python
import pandas as pd
import geopandas

import contextily
```

```python
def add_geometry(df):
   gdf = geopandas.GeoDataFrame(df,
            geometry=geopandas.points_from_xy(df["lon"],df["lat"]),
            crs="EPSG:4326")

   return gdf
```

```python
def read_data(filename):
    df  = pd.read_csv(filename,sep=',')
    df['start_date'] = pd.to_datetime(df['start_date'])
    df['end_date'] = pd.to_datetime(df['end_date'])
    gdf = add_geometry(df)
    return gdf
```

```python
def select_metropole(df):
    sub_df = df[(df['lon']>=-5.2) & (df['lon']<=9.8) &
                (df['lat']>=41.3) & (df['lat']<=51.3)]
    return sub_df
```

```python
csv_file = 'data/stations-meteo-france.csv'

stations = read_data(csv_file)
stations.shape
```

```python
stations.head()
```

```python
metropole = select_metropole(stations)
metropole.shape
```

```python
nb_stations_by_type = metropole.groupby(by='station_type').count()
nb_stations_by_type
```

```python
stations_at_Jp1 = metropole[metropole['station_type']<=2]
print(len(stations_at_Jp1))

stations_at_Jp1 = stations_at_Jp1.to_crs("EPSG:3857")
ax = stations_at_Jp1.plot(marker='o', color='black', markersize=3)
contextily.add_basemap(ax, source=contextily.providers.OpenStreetMap.Mapnik)
```

```python
stations_at_Jp21 = metropole[(metropole['station_type']==3) | (metropole['station_type']==4)]

stations_at_Jp21 = stations_at_Jp21.to_crs("EPSG:3857")
ax2 = stations_at_Jp1.plot(marker='o', color='black', markersize=3)
stations_at_Jp21.plot(ax=ax2,marker='o', color='red', markersize=3,alpha=0.5)
contextily.add_basemap(ax2, source=contextily.providers.OpenStreetMap.Mapnik)
```
